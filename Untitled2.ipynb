```python
HH = zeros((model.nNodes, 2))
func = lambda x, a, b, c, d, e, f, g:  b * exp(-c * (x) )   +  d * exp(-e * (x-f)) 

colors = cm.tab20(arange(model.nNodes))
xx = linspace(0, 50, 10000)
g = gp(normalize_y = True)
fig, ax = subplots(figsize = (10, 10))
for idx, y in enumerate(tqdm(mi.T)):
    x = arange(len(y))
    a, b = scipy.optimize.curve_fit(func, x, y, maxfev = 10000000)
    f = lambda x, p, c: func(x, *p) - c
    tmp  = .25 * max(y)
#     tmp  = .5
    
#     g.fit(x[:,None], y)
    
    ff = lambda x, c : g.predict(x[:, None]) - c
#     print(tmp)
    r = scipy.optimize.root(f, 0, args = (a, tmp) )
    rr = scipy.optimize.root(f, 0, args = (a, .05), tol = .1)
#     rrr = scipy.optimize.root(ff, 0, args = (a[-1] + .1))
    print(a[-1], rrr.x)
    HH[idx, :] = r.x, rr.x
    
    ax.scatter(rrr.x, ff(rr.x, 0))
    
#     ax.plot(xx, g.predict(xx[:,None]))
    ax.plot(xx, func(xx, *a), color = colors[idx], alpha = .2)
    
    ax.scatter(r.x, func(r.x, *a), s = 120, color = colors[idx], marker = '^')
    ax.scatter(rr.x, func(rr.x, *a), s = 60, color =  colors[idx], marker = 's', label = idx)
    ax.scatter(x, y, color = colors[idx], alpha = .4, marker = 'o')
    
    error = sqrt((func(x, *a) - y)**2).mean()
    ax.set_title(error)
# ax.legend()
    
    ax.set_xlim(0, 50)
show()
```

```python
def tsplot(y, er, x = None):
    if x is None:
        x = arange(len(y))
    fig, ax = subplots();
    for xi, ei in zip(y, er):
        ax.plot(xi)
        ax.fill_between(arange(len(xi)), xi + ei,  xi-ei, alpha = .3)

degs = dict(nx.degree(model.graph))
tmp = {}
for node, d in degs.items():
    idx = model.mapping[node]
    tmp[d] = tmp.get(d, ()) + (ii[idx, 0],)

xx = zeros(len(tmp))
yy = zeros(len(tmp))
for idx, (k, v) in enumerate(tmp.items()):
    xx[idx] = mean(v)
    yy[idx] = std(v) 
xxx = array(list(tmp.keys()))
fig, ax = subplots()
ax.errorbar(xxx, xx, yerr = yy, fmt = 'o')
ax.set_yscale('log')

fig, ax = subplots()
ax.scatter(xxx, xx)
for node, deg in degs.items():
    idx = model.mapping[node]
    ax.scatter(deg, ii[idx, 0])
ax.set_yscale('log')
```

```python

degs = dict(nx.degree(model.graph))
idt_deg = {}
idt_norm = {}
idt_rel  = {}
idt_rel_n= {}
for node, deg in degs.items():
    idx = model.mapping[node]
    idt_deg[deg] = idt_deg.get(deg, 0) + ii[idx, 0]
    idt_norm[deg] = idt_norm.get(deg, 0) + 1
    
    idt_rel_n[deg] = idt_rel.get(deg, 0) + 1
    idt_rel[deg] = idt_rel.get(deg, 0) + ii[idx, 1]

idts = {}
for k, v in idt_deg.items():
    idts[k] = v / idt_norm[k]
for k, v in idt_rel.items():
    idt_rel[k] = v / idt_rel_n[k]

#     ax.scatter(deg, ii[idx, 1])
fig, ax = subplots();
ax.scatter(list(idts.keys()), log(list(idts.values())))
setp(ax, **dict(xlabel = 'degree', ylabel = 'abs idt'))
fig, ax = subplots()
ax.scatter(list(idt_rel.keys()), list(idt_rel.values()))
print(sort(list(idts.values())))
setp(ax, **dict(xlabel = 'degree', ylabel = 'rel idt'))

ii[ii <0]= 0
# ax.set_yscale('log')
# ax.set_xlim(0, 30)
# ax.set_xscale('log')
fig, ax = subplots()
ax.hist(ii[:, 1])
# ax.set_yscale('log')
```
