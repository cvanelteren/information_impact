```python
import numpy as np, holoviews as hv
hv.extension('bokeh')
data = np.random.rand(100, 2)
scatter = hv.Scatter(data)

# for stream plots the label has to be pre-determined
stream  = hv.streams.Selection1D(source = scatter)
sampled = hv.NdOverlay({\
           i : hv.Curve(np.random.rand(100,2)) for i in range(len(data))\
          })
def update(index):
    if not index:
        return sampled[0].relabel('No selection')
    return sampled[index[0]]
dm = hv.DynamicMap(update, streams = [stream])
layout = scatter + dm
layout.opts(hv.opts.Scatter(size = 10, tools = 'tap hover'.split(), framewise = False))

```

```python
from bokeh.layouts import column
from bokeh.models import CustomJS, ColumnDataSource, Slider
from bokeh.plotting import Figure, output_file, show, output_notebook
import bokeh, bokeh.plotting as bp
output_notebook()

# create two datasets 
tmp = dict(x = np.random.rand(100), \
           y = np.random.rand(100))

# scatter plot to use for source selection
s1 = bokeh.models.ColumnDataSource(data = tmp)

# define empty buffer

tmp = dict(x = [], y = [])
s2 = bokeh.models.ColumnDataSource(data = tmp, name = 'test')


# create mock datasets
x = np.arange(100)
tmp = dict(\
#            x =[[x for _ in range(5)] for i in range(100)],\
           y = np.random.rand(100, 5, 100).tolist(),
#            y =[[np.exp(-x * np.random.rand()) for _ in range(5)] for i in range(100)]\
          )

ty = [[np.exp(-x * np.random.rand()) for _ in range(5)] for i in range(100)]
print(np.array(ty).shape)
s3 = bokeh.models.ColumnDataSource(tmp)
callback = """
var inds = cb_obj.indices;
var aucs   = s1.data;
var buffer = s2.data;
var mi     = s3.data;

buffer['x']  = []; 
buffer['y']  = [];

var x, y;
var xx = [], yy = [];

var I = inds.length;
var idx, J;
for (var i = 0; i < I ; i++)
    {
    idx = inds[i];
    N   = mi['y'][idx].length;
    for (var j = 0 ; j < N ; j++){
        xx = [];
        J  = mi['y'][idx][0].length;
        for (var k = 0; k < J ; k++){
            xx.push(k);
        }
        yy = mi['y'][idx][j];
        buffer['x'].push(xx);
        buffer['y'].push(yy)
        
    }
    
    }
s2.change.emit();
"""


left  = bp.figure(tools = 'hover tap lasso_select'.split(), width = 300)
right = bp.figure()

left.scatter('x', 'y', source = s1, size = 10)
right.multi_line('x', 'y', source = s2)

# add the callback to the plot
s1.selected.js_on_change('indices', bokeh.models.CustomJS(\
                                                          args = dict(s1 = s1, s2 = s2, s3 = s3),\
                                                          code = callback))
show(bokeh.layouts.row(left, right))
```
