# Value chains modelling 
In order to model value chain we apply a statistical physics approach. Given a graph $G = (V,E)$ we assign each node $v\in V$ with dynamic a possible rule alphabet of size $X$ according to a transition rule $R = (V_r, E_r, W_r)$. The master equation will be modelled using the Gibbs measure for observing the a node in some state $x\in X$. 
$$ P_{v_i}(X = x) = \frac{\exp( \beta e(x))}{\sum_x \exp(\beta e(x))}$$

with $$e(x) = \sum_{<i,j> \in G} W(x_i, x_j)$$
where $W_{ij}$ is the edge weight in the rule graph $R$. Transition between states will follow the Metropolis-Hasting algorithm:

$$ P_{x \rightarrow x'} = \exp( -\beta  (e(xÂ´) - e(x))$$


## Value chains as circular chain
The circular chain has rules in a configuration 

$$ a_n \leftrightarrow ... \leftrightarrow a_1 \leftrightarrow a_0 \leftrightarrow ... \leftrightarrow a_n$$


## CASE: Chain graph
We take here a simple path graph of size $N$ and vary the size of the rule-ring shown above and observe what happens. Each element in the system will be satisfied if it is is neighboring an edge in the rule graph, every other combination will be set to $-1$ energy. Update happens according to the Gibbs measure.

```python
import networkx as nx, numpy as np, plotly.graph_objects as go
from plexsim.models import *

def generate_cycle(n, self_edge = False, weight = 1) -> nx.graph:
    """
    Generates local cycle where only the edges are considered to be 1 otherwise 0
    Input:
    :n: integer representing number of nodes in the system.
    
    Returns:
    :nx.Graph: with the edges of a cycle graph as 1 else 0
    """
    g = nx.complete_graph(n)
    nx.set_edge_attributes(g, -weight, 'weight')
    for u, v in nx.path_graph(n).edges():
        g[u][v]['weight'] = 1
    
    if self_edge:
        for u in g.nodes():
            g.add_edge(u, u, weight = weight)
    return g

# show example rule graph 
ni = 3 #number of states
rule_graph = generate_cycle(ni, False, 1)
hm  = go.Heatmap(z = nx.adj_matrix(rule_graph).todense(),\
                colorbar = dict(title = 'Weight'))
fig = go.Figure(data = hm)
fig.update_layout(title = f'{ni}-state rule graph', \
                  autosize = False,\
                  width = 600,\
                  xaxis_title = 'State',\
                  yaxis_title = 'State',\
                )
fig.show()
```

Example of a rule-set. Weights are given in the colorbar.

# Order parameter
Change the order parameter and check for the average phase of the function:
- As a function of 

```python
%%capture
n = 16
# g = nx.grid_graph([n, n])
g = nx.path_graph(10)
rules = np.arange(2, 15, dtype = float)

temps  = np.linspace(0, 5, 30, dtype = float)
trials = 10
weight = 1
orderData = np.zeros((rules.size, trials, temps.size, 2))

for idx, rule in enumerate(rules):
    tmp = generate_cycle(int(rule), False, weight = weight)
    AS = np.arange(rule)
    m = Potts(graph = g, rules = tmp, agentStates = AS)
    for trial in range(trials):
        orderData[idx, trial, :] = m.magnetize(temps, \
                                   n = int(1e2)).T
```

```python


from scipy.stats import sem
lines = []
for idx, rule in enumerate(rules):
    line = go.Scatter(x = temps,\
                      y = orderData[idx,..., 0].mean(0),
                      error_y = dict(array = sem(orderData[idx, ..., 0], 0) * 2),\
                      name = rule,\
                      mode = 'lines')
    lines.append(line)
fig = go.Figure(data = lines)
fig.update_layout(\
                 xaxis_title = r'$\theta$',\
                 yaxis_title = 'phase',\
#                  yaxis = dict(dtick = "log", type = "log"),\
                 width = 600,\
                 autosize = False,\
                 legend_title = '$|R|$')
fig.show()
```

For fixed system size, there exists a notable increase in the aligned spins as a function of chain length. The system seems to stabilize as the value chain increases. 

```python
# plot averages of populations over time
sim = m.simulate(100)
lines = []
for state in range(ni):
    line = go.Scatter( y = (sim == state).mean(1), \
#                       legendgroup = 'state', \
                      name = f'{state}')
    lines.append(line)
lines.append(\
     go.Scatter(y = sim.mean(1), \
                name = 'Average of system',\
                yaxis = 'y2')
            )
fig = go.Figure(data = lines)
fig.update_layout(\
                 xaxis_title = 'time',\
                 yaxis_title = 'state mean',\
                 legend_title = 'State',\
                 yaxis2 = dict(overlaying = 'y',
                              side = 'right', \
                              title = 'system mean'))
fig.show()
```

```python
def checkVC(data, model):
    results = np.zeros((data.shape))
    for idx, state in enumerate(data):
        for node in range(model.nNodes):
            neighbors = np.array([ i for i in model.adj[node]['neighbors'].keys()])
            tmp = set([int(i) for i in state[[*neighbors]]])
            r   = set(m.rules.neighbors(state[node])) # will not work for all  
            N = 0 
            for neighbor in m.rules.neighbors(state[node]):
                weight = m.rules[state[node]][neighbor]['weight']
                if weight > 0:
                    N +=  1
            results[idx, node] = len(r.intersection(tmp)) == N
    return results
# res = checkVC(tmp, m)
```

```python
import plotly.graph_objects as go
from jupyter_plotly_dash import JupyterDash
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.colors as plc

class ModelVisualizer(JupyterDash):
    def __init__(\
                self,\
                model, 
                layout,\
                name = ''):
        # init class 
        super(ModelVisualizer, self).__init__(name)
        
        # create figure
        self.fig = self.create_figure(model, layout)
        # remove axes
        self.fig.update_layout(xaxis = dict(showticklabels = False),\
                              yaxis = dict(showticklabels = False))
        # init app layout 
        layout = [dcc.Graph(id = 'graph', figure = self.fig)]
        self.layout = html.Div(layout)
        
        self.model = model
     
    def create_figure(self, model, layout):
        """
        Converts networkx graph to plotly figure structure
        """
        # todo give properties to nodes in layout or in separate?
        nodes = []
        edges = []
        
        # add nodes
        for node, (x, y) in layout.items():
            node = go.Scatter(x = [x], y = [y], mode = 'markers',\
                             marker = dict(\
                                       color = 'cornflowerblue',\
                                       size  = 10,\
                                              ),\
                              name = model.mapping[node],\
                              showlegend = False,
                             )
            nodes.append(node)
            
        # add edges 
        for u, v in model.graph.edges():
            x1, y1 = layout.get(u)
            x2, y2 = layout.get(v)
            edge = go.Scatter(x = [x1, x2], \
                              y = [y1, y2], mode = 'lines',\
                             line = dict(color = 'black'),
                             showlegend = False)
            edges.append(edge)
        return go.Figure(data = [*edges, *nodes])
    
    def create_slider(self):
        for div in self.layout:
            print(div)
            if div == 'time-slider':
                return div
            
        slider = dcc.Slider(
                        id   = 'time-slider',
                        min  = 0,
                        max  = 1,\
                        step = 1,
                        value= 0,
                        updatemode =  'drag')
        
        self.layout.children.append(slider)
        return slider
    
    def setup_colors(self, n: int):
        colors = plc.sequential.Viridis
        cdx = np.linspace(0, 1, len(colors))
        idx = np.linspace(0, len(colors) - 1, n, dtype = int)
        return [colors[i] for i in idx]
        
    def show_simulation(self, data):
        #obtain references
        nodes = {}
        for go in self.fig.data:
            try:
                if int(go.name) in self.model.rmapping:
                    nodes[int(go.name)] = go
            except:
                continue
        
        T, N = data.shape
        # setup slider
        slider = self.create_slider() 
        # reset slider
        slider.value = 0
        slider.min   = 0 
        slider.max   = T - 1
        
        colors = self.setup_colors(self.model.nStates)
#         print(len(colors), colors)
        @self.callback(Output('graph', 'figure'),\
                       [Input('time-slider', 'value')],\
                       )
        def simulation_callback(value):
            for node in self.model.rmapping:
                coloridx = int(data[value, node])
#                 c = np.random.choice(colors)
                nodes[node].marker.color = colors[coloridx]
#                 nodes[node].marker.color = c
            return self.fig 
        return self
                
        
        
        
        # setup callback
# g = nx.grid_graph([4, 4])
N   = 4


rules = generate_cycle(N, True, weight = 1)
g = nx.star_graph(5)
g = nx.path_graph(5)
m = Potts(g, rules =  rules, \
          t = 3.5, \
          agentStates = np.arange(N)) 

lay = nx.kamada_kawai_layout(m.graph)
mv = ModelVisualizer(m, lay)
simt = m.simulate(1000)

# res = checkVC(simt, m)
# line = go.Heatmap(z = res.T)
# fig = go.Figure(data = line)
# fig.show()

mv.show_simulation(simt)
```
