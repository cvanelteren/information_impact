# Value chains modelling 
In order to model value chain we apply a statistical physics approach. Given a graph $G = (V,E)$ we assign each node $v\in V$ with dynamic a possible rule alphabet of size $X$ according to a transition rule $R = (V_r, E_r, W_r)$. The master equation will be modelled using the Gibbs measure for observing the a node in some state $x\in X$. 
$$ P_{v_i}(X = x) = \frac{\exp( \beta e(x))}{\sum_x \exp(\beta e(x))}$$

with $$e(x) = \sum_{<i,j> \in G} W(x_i, x_j)$$
where $W_{ij}$ is the edge weight in the rule graph $R$. Transition between states will follow the Metropolis-Hasting algorithm:

$$ P_{x \rightarrow x'} = \exp( -\beta  (e(xÂ´) - e(x))$$


## Value chains as circular chain
The circular chain has rules in a configuration 

$$ a_n \leftrightarrow ... \leftrightarrow a_1 \leftrightarrow a_0 \leftrightarrow ... \leftrightarrow a_n$$


## CASE: Chain graph
We take here a simple path graph of size $N$ and vary the size of the rule-ring shown above and observe what happens. Each element in the system will be satisfied if it is is neighboring an edge in the rule graph, every other combination will be set to $-1$ energy. Update happens according to the Gibbs measure.

```python
import networkx as nx, numpy as np, plotly.graph_objects as go
from plexsim.models import *

def generate_cycle(n, self_weight = 0, weight = 1) -> nx.graph:
    """
    Generates local cycle where only the edges are considered to be 1 otherwise 0
    Input:
    :n: integer representing number of nodes in the system.
    
    Returns:
    :nx.Graph: with the edges of a cycle graph as 1 else 0
    """
    g = nx.complete_graph(n)
    nx.set_edge_attributes(g, -weight, 'weight')
    for u, v in nx.path_graph(n).edges():
        g[u][v]['weight'] = 1
    
    for u in g.nodes():
        g.add_edge(u, u, weight = self_weight)
    return g

# show example rule graph 
ni = 3 #number of states
rule_graph = generate_cycle(ni, False, 1)
# rule_graph = nx.turan_graph(5, 3)
hm  = go.Heatmap(z = nx.adj_matrix(rule_graph).todense(),\
                colorbar = dict(title = 'Weight'))
fig = go.Figure(data = hm)
fig.update_layout(title = f'{ni}-state rule graph', \
                  autosize = False,\
                  width = 600,\
                  xaxis_title = 'State',\
                  yaxis_title = 'State',\
                )
fig.show()
```

Example of a rule-set. Weights are given in the colorbar.

```python
%%capture
n = 16
# g = nx.grid_graph([n, n])
g = nx.path_graph(10)
rules = np.arange(2, 15, dtype = float)

temps  = np.linspace(0, 5, 30, dtype = float)
trials = 10
weight = 1
orderData = np.zeros((rules.size, trials, temps.size, 2))

for idx, rule in enumerate(rules):
    tmp = generate_cycle(int(rule), False, weight = weight)
    AS = np.arange(rule)
    m = Potts(graph = g, rules = tmp, agentStates = AS)
    for trial in range(trials):
        orderData[idx, trial, :] = m.magnetize(temps, \
                                   n = int(1e2)).T
```

```python


from scipy.stats import sem
lines = []
for idx, rule in enumerate(rules):
    line = go.Scatter(x = temps,\
                      y = orderData[idx,..., 0].mean(0),
                      error_y = dict(array = sem(orderData[idx, ..., 0], 0) * 2),\
                      name = rule,\
                      mode = 'lines')
    lines.append(line)
fig = go.Figure(data = lines)
fig.update_layout(\
                 xaxis_title = r'$\theta$',\
                 yaxis_title = 'phase',\
#                  yaxis = dict(dtick = "log", type = "log"),\
                 width = 600,\
                 autosize = False,\
                 legend_title = '$|R|$')
fig.show()
```

For fixed system size, there exists a notable increase in the aligned spins as a function of chain length. The system seems to stabilize as the value chain increases.

```python
def checkVC(data, model):
    results = np.zeros((data.shape))
    for idx, state in enumerate(data):
        for node in range(model.nNodes):
            neighbors = np.array([ i for i in model.adj[node]['neighbors'].keys()])
            tmp = set([int(i) for i in state[[*neighbors]]])
            r   = set(m.rules.neighbors(state[node])) # will not work for all  
            N = 0 
            for neighbor in m.rules.neighbors(state[node]):
                weight = m.rules[state[node]][neighbor]['weight']
                if weight > 0:
                    N +=  1
            results[idx, node] = len(r.intersection(tmp)) == N
    return results

def checkEnergy(data, model):
    results = np.zeros(data.shape)
    for idx, states in enumerate(data):
        tmp = model.siteEnergy(states)
        results[idx] = tmp 
    return results
# res = checkVC(tmp, m)
```

```python
import plotly.graph_objects as go
from jupyter_plotly_dash import JupyterDash
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.colors as plc

class ModelVisualizer(JupyterDash):
    def __init__(\
                self,\
                model, 
                layout,\
                name = ''):
        # init class 
        super(ModelVisualizer, self).__init__(name)
        
        # create figure
        self.fig = self.create_figure(model, layout)
        # remove axes
        self.fig.update_layout(xaxis = dict(showticklabels = False),\
                              yaxis = dict(showticklabels = False))
        # init app layout 
        layout = [dcc.Graph(id = 'graph', figure = self.fig)]
        self.layout = html.Div(layout)
        
        self.model = model
     
    def create_figure(self, model, layout):
        """
        Converts networkx graph to plotly figure structure
        """
        # todo give properties to nodes in layout or in separate?
        nodes = []
        edges = []
        
        # add nodes
        for node, (x, y) in layout.items():
            node = go.Scatter(x = [x], y = [y], mode = 'markers',\
                             marker = dict(\
                                       color = 'cornflowerblue',\
                                       size  = 10,\
                                              ),\
                              name = model.mapping[node],\
                              showlegend = False,
                             )
            nodes.append(node)
            
        # add edges 
        for u, v in model.graph.edges():
            x1, y1 = layout.get(u)
            x2, y2 = layout.get(v)
            edge = go.Scatter(x = [x1, x2], \
                              y = [y1, y2], mode = 'lines',\
                             line = dict(color = 'black'),
                             showlegend = False)
            edges.append(edge)
        return go.Figure(data = [*edges, *nodes])
    
    def create_slider(self):
        for div in self.layout:
            print(div)
            if div == 'time-slider':
                return div
            
        slider = dcc.Slider(
                        id   = 'time-slider',
                        min  = 0,
                        max  = 1,\
                        step = 1,
                        value= 0,
                        updatemode =  'drag')
        
        self.layout.children.append(slider)
        return slider
    
    def setup_colors(self, n: int):
        colors = plc.sequential.Viridis
        cdx = np.linspace(0, 1, len(colors))
        idx = np.linspace(0, len(colors) - 1, n, dtype = int)
        return [colors[i] for i in idx]
        
    def show_simulation(self, data):
        #obtain references
        nodes = {}
        for go in self.fig.data:
            try:
                if int(go.name) in self.model.rmapping:
                    nodes[int(go.name)] = go
            except:
                continue
        
        T, N = data.shape
        # setup slider
        slider = self.create_slider() 
        # reset slider
        slider.value = 0
        slider.min   = 0 
        slider.max   = T - 1
        
        colors = self.setup_colors(self.model.nStates)
#         print(len(colors), colors)
        @self.callback(Output('graph', 'figure'),\
                       [Input('time-slider', 'value')],\
                       )
        def simulation_callback(value):
            for node in self.model.rmapping:
                coloridx = int(data[value, node])
#                 c = np.random.choice(colors)
                nodes[node].marker.color = colors[coloridx]
#                 nodes[node].marker.color = c
            return self.fig 
        return self
                
        
        
        
        # setup callback
# g = nx.grid_graph([4, 4])
N   = 2
rules = generate_cycle(N, True, weight = 0)
# g = nx.path_graph(50)
g = nx.ladder_graph(50)
# g = nx.grid_graph([4, 4])
# g = nx.erdos_renyi_graph(10, .3)
# g = nx.cycle_graph(10)
m = Potts(g, \
          rules =  rules, \
          t = 0.2,\
          agentStates = np.arange(N),\
#          memento = 1,\
         ) 

lay = nx.kamada_kawai_layout(m.graph)
mv = ModelVisualizer(m, lay)
simt = m.simulate(100)

res = checkEnergy(simt, m)
line = go.Heatmap(z = res.T)
fig = go.Figure(data = line)
fig.show()

mv.show_simulation(simt)
```

```python

"""
This example demonstrates how to visualise a NetworkX graph using a VisPy
Graph.
"""
%load_ext autoreload
%autoreload 2
import sys


from vispy import app, scene
from vispy.visuals.graphs import layouts
from vispy.visuals.graphs.util import _straight_line_vertices, issparse
import networkx as nx, numpy as np
from imi.utils.graph import recursive_tree

class NetworkxCoordinates:
    def __init__(self, graph, layout = None, *args, **kwargs):
        self.graph = graph
        self.positions = np.zeros((len(graph), 2), dtype = np.float32)
        # default random positions
        if type(layout) is type(None):
            self.positions = np.random.rand(*self.positions.shape)
            
        # check for networkx
        elif isinstance(layout, str):
            layout += "_layout" # append for nx
            if f := getattr(nx, layout):
                self.positions = np.asarray([i for i in dict(f(graph, **kwargs)).values()])
            else:
                raise ValueError("Check networkx for layouts")
        # assume dict from networkx; values are 2-array
        elif isinstance(layout, dict):
            self.positions = np.asarray([i for i in layout.values()])
            
        # assume given values
        elif isinstance(layout, np.ndarray):
            assert layout.ndim == 2
            assert layout.shape[0] == len(graph)
            self.positions = layout
        else:
            raise ValueError("Input not understood")
            
        # normalize coordinates
        self.positions = (self.positions - self.positions.min()) / (self.positions.max() - self.positions.min())
        self.positions = self.positions.astype(np.float32)
            
    def __call__(self, adjacency_mat, directed = False):
        if issparse(adjacency_mat):
            adjacency_mat = adjacency_mat.tocoo()
        line_vertices, arrows = _straight_line_vertices(adjacency_mat, self.positions, directed)
        
        yield self.positions, line_vertices, arrows
        
    @property
    def adj(self):
        return nx.adjacency_matrix(self.graph)

# g = recursive_tree(5, 1)
from imi.utils.plotting import get_shell_layout
# p = get_shell_layout(g, rotate = -.1)

g = nx.erdos_renyi_graph(100, .2)
layout = NetworkxCoordinates(graph = g, layout = "spring")

# create canvas
canvas = scene.SceneCanvas(title='Simple NetworkX Graph', size=(600, 600),
                           bgcolor='black', show=True)
view = canvas.central_widget.add_view('panzoom')
visual = scene.visuals.Graph(
    layout.adj, layout= layout, line_color= (1, 1, 1,  .5), arrow_type="stealth",
    arrow_size=30, node_symbol="disc", node_size=20,
    face_color=(1, 0, 0, 1), border_width=0.0, animate=False, directed=False,
    parent=view.scene)


@canvas.events.draw.connect
def on_draw(event):
    if not visual.animate_layout():
        canvas.update()
    
canvas.show()
```

```python
from objects import properties, self_properties


class Car:
    with properties(locals(), 'meta') as meta:

        @meta.prop(read_only=True)
        def brand(self) -> strt:
            """Brand"""

        @meta.prop(read_only=True)
        def max_speed(self) -> float:
            """Maximum car speed"""

        @meta.prop(listener='_on_acceleration')
        def speed(self) -> float:
            """Speed of the car"""
            return 0  # Default stopped

        @meta.prop(listener='_on_off_listener')
        def on(self) -> bool:
            """Engine state"""
            return False

    def __init__(self, brand: str, max_speed: float = 200):
        self_properties(self, locals())

    def _on_off_listener(self, prop, old, on):
        if on:
            print(f"{self.brand} Turned on, Runnnnnn")
        else:
            self._speed = 0
            print(f"{self.brand} Turned off.")

    def _on_acceleration(self, prop, old, speed):
        if self.on:
            if speed > self.max_speed:
                print(f"{self.brand} {speed}km/h Bang! Engine exploded!")
                self.on = False
            else:
                print(f"{self.brand} New speed: {speed}km/h")
        else:
            print(f"{self.brand} Car is off, no speed change")
```
