#### Value chains
We hadden vorige keer besproken dat dit mogelijk handige stap was omdat het in lijn is met Paul's onderzoek. Het is vanuit onze kant dan een stap in de richting van de blackbox van crimineel gedrag zoals hierboven beschreven. 

Het idee was om een simpele (conceptuele) value chain te pakken. Bv het process waarbij rollen $A,B,C$ de volgende relatie hebben:

$$ A \rightarrow B \rightarrow C$$

Dit zouden we kunnen doen door de hamiltonian van het Potts model aan te passen

$$ H(\sigma_i) = \sum_{ij} [\cos \Delta \theta_{ij} + f(\Delta \theta_{ij})]$$
waarbij $f(\theta_{ij})$ de chain moet encoderen. Ik stel voor het nu dit een delta functie te laten: $f(\Delta \theta_{ij}) = Z\delta( \theta_i - \theta_j)$ waarbij $Z =  (1 - \cos(\theta_i - \theta_j))$ de functie schaalt opdat de relatie van $H$ per node nog steeds binnen $[-1,1]$ schaalt.

### Nudge interventions
Voor causale interventies is het cruciaal om een helder algemeen framework te hebben waarin via contact processen (hierboven) te kunnen doen. Ik stel voor dat we het concept van de node toevoegen anders definieren. Laat $\eta_i \sim \textrm{binom(p)}$ de externe interventie zijn met success / fail rate. Tevens laat een element in het netwerk (node) random variabele zijn $\sigma_i \sim P(\sigma_i = x)$ volgens een bepaalde verdeling $(P(\sigma_i = x)$. Bij een succesvolle interventie zal $\eta_i \rightarrow \sigma_i^{t+1} = \textrm{uniform}(\sigma_i)$, i.e. de interventie zorgt ervoor dat de state van een node random gekozen wordt over de mogelijke states van die node. Voor een continue definitie had ik een idee om iets via de characteristic equation te definieren met een soort fase verandering binnen het frequentie spectrum (bv dmv 2e moment te pakken). Een ander idee was via een soort max entropy principe icm de momenten en variante verandering met norm 0 zoals Rick voorstelde. Dit zou je kunnen voorstellen als een 'difussie' process op een kansverdeling waarbij de PDF / PMF gelijk verdeeld wordt en proportioneel met $\eta$ verandert. 

Ik weet niet of we dit zouden kunnen benaderen analytisch om te laten zien dat er een soort optimale nudge is opdat de informatie transmissie optimaal is, zoals in https://www.nature.com/articles/srep01223.

De definitie hierboven is volgens mij toepasbaar op welke dynamiek zonder dat daar een distributie van bekend hoeft te zijn. 

### Operationalisering van paper 1
Korte toelichting over toepassen van interventions op dynamiek van RBN en SIR. 
#### RBN
- Initialiseer random regels op graaf $G$;
- loop voor $T$ tijdstappen.

Herhaal $N$ keer. Interventies kunnen op elke node op elk moment ingezet worden en zorgen voor een state flip.

### SIR
- Infecteer 1 node random op graaf $g$;
- Loop for $T$ tijdstappen.

Herhaal $N$ keer.  Interventies kunnen op elke node op elk moment ingezet worden en zorgen voor een state flip.
