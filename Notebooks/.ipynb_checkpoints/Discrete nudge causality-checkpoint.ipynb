## Does not work

Maybe it was a foolish thought in hindsight, but no this does not work in this way, because of course the reversed system $(Y,X)$ is also a noisy channel, so doing a 0.01 nudge on the Y results in less than 0.01 impact on X, which if I would do that to the original system $(X,Y)$ I would not get a 0.01 impact on Y.

# Discover the 'efficiency' of the causal relationship

The idea here is to find the 'efficiency' of a causal relationship $X \rightarrow Y$, such that we know more or less how strong the nudge on $X$ has to be so that we get a chosen (say 0.01) norm of impact on $Y$.

```python
pdf = jp.JointProbabilityMatrix(2, 5)  # new pdf
old_pdf = pdf.copy()  # keep original since pdf.nudge() changes in-place
```

## Analytical considerations regarding $||i_b||$ as function of $||\vec{\epsilon}||$

The goal here is to verify that if we utilize a small norm of $||\vec{\epsilon}||$ that then $||\vec{i}||$ (or rather $||\vec{i_b / p(b)}||$) is also small, as required by our analytical derivations.

Here the channel is $A \rightarrow B$. $p'(b)$ is the probability of $B=b$ after the nudge; $p(b)$ is before.

$$p'(b) = \sum_a p'(a)p(b|a) = \sum_a p(a)p(b|a) + \sum_a \epsilon_a p(b|a) = p(b) + \sum_a \epsilon_a p(b|a)$$
$$i_b := p'(b) - p(b) = \sum_a \epsilon_a p(b|a) = f(\vec{\epsilon})$$

If we consider the elements of the vector $\vec{\epsilon_a}$ as samples from a stochastic variable, and likewise for $\vec{p(b)}$ across the values $b$ (which, as function of different choices of $\epsilon$, can be considered as delta functions), then we could interpret the summand as a product distribution: https://en.wikipedia.org/wiki/Product_distribution. Observe that the two 'stochastic variables' are independent: the $\epsilon$ vector is drawn independently from $p(b|a)$.

In this case the expected value (over different $\epsilon$) equals apparently: 

$$\left\langle i_b \right\rangle_{\epsilon} = \sum_a \left\langle \epsilon_a \right\rangle_{\epsilon} \cdot \left\langle p(b|a) \right\rangle_{\epsilon} = \sum_a 0 \cdot p(b|a) = 0.$$

In our analytical derivation the term $\frac{i_b^2}{p(b)}$ actually is important, so let's look at its distributional properties.

$$i_b^2 = \left( \sum_a \epsilon_a \cdot p(b|a) \right)^2 = \sum_{a,a'} \epsilon_a \epsilon_{a'} \cdot p(b|a) p(b|a').$$

If we take the expectation then the 'cross terms' ($a \neq a'$) vanish:

$$\left\langle i_b^2 \right\rangle_{\epsilon} = \sum_{a,a'} \left\langle \epsilon_a \epsilon_{a'} \cdot p(b|a) p(b|a') \right\rangle_{\epsilon} = \sum_{a} \left\langle \epsilon_a^2 \cdot p(b|a)^2 \right\rangle_{\epsilon}.$$

Taking again the road of the product distribution, this time each term in the sum becomes:

$$\left\langle i_b^2 \right\rangle_{\epsilon} = \sum_{a} \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot \left\langle p(b|a)^2 \right\rangle_{\epsilon} = \sum_{a} \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot p(b|a)^2.$$

Let's see if we can make it a bit nicer. Using Bayes' theorem here: $p(b) = \frac{p(b|a)p(a)}{p(a|b)}$.

[comment]: # ($$\left\langle i_b^2 \right\rangle_{\epsilon} = \sum_{a} \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot p(b|a)^2 = \sum_{a} \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot p(b|a)^2 \frac{p(a)}{p(a)} = \sum_{a} p(a) \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot p(b|a)^2 \frac{p(b|a)}{p(a|b)p(b)} = \sum_{a} p(a) \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot \frac{p(b|a)^3}{p(a|b)p(b)}.$$)

$$\left\langle i_b^2 \right\rangle_{\epsilon} = \sum_{a} \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot p(b|a)^2 = \sum_{a} \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot p(b|a)^2 \frac{p(b)}{p(b)} = \sum_{a} p(b) \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot p(b|a)^2 \frac{p(a|b)}{p(b|a)p(a)} = \sum_{a} p(b) \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot \frac{p(a|b)p(b|a)}{p(a)} = \sum_{a} p(b) \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot \frac{p(b|a)p(b|a)}{p(b)} = \sum_{a} \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot p(b|a)^2.$$

Ok that didn't help.

Since each element of $\vec{\epsilon}$ is statistically indistinguishable, their expectation does not depend on $a$ and can be taken out.

$$\left\langle i_b^2 \right\rangle_{\epsilon} = \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot \sum_{a} p(b|a)^2.$$

[comment]: # (This is for a particular $b$. Observe that for a particular $b$ this impact can grow with the number of states $|A|=n$, since worst-case each $p(b|a)=1$.)

[comment]: # ($$\left\langle i_b^2 \right\rangle_{\epsilon} = \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot n \cdot \sum_{a=1}^n \frac{1}{n} p(b|a)^2.$$)

Let us now consider the expectation over $p(b)$ of both sides (denoted $\left\langle . \right\rangle_{b}$):

$$\left\langle i_b^2 \right\rangle_{\epsilon,b} = \left\langle \epsilon_a^2 \right\rangle_{\epsilon} \cdot \sum_{a} \left\langle p(b|a)^2 \right\rangle_b.$$

Observe that $\left\langle p(b|a)^2 \right\rangle \leq 1$ which can be achieved by setting a single $p(b|a)=1$ and the rest zero, making also the corresponding $p(b)=1$. This means that $\sum_{a} \left\langle p(b|a)^2 \right\rangle \leq n$ since $n=|A|$. This means that, worst-case, the expected impact can be $n$ times larger than the average nudge size. ($\sqrt{n}$ times larger in case we consider Euclidea norms instead.) The minimum occurs for the uniform distribution ($p(b|a)=1/n$): $\sum_{a} \left\langle p(b|a)^2 \right\rangle = n \frac{1}{n^2} \geq 1/n$. So:

$$1/n \leq \frac{\left\langle i_b^2 \right\rangle_{\epsilon,b}}{\left\langle \epsilon_a^2 \right\rangle_{\epsilon}} \leq n.$$

The following step I am doing a bit intuitively, hopefully it is correct. Since both $i_b$ and $\epsilon_a$ have expectation zero, the two second moments are equal to their variance. Therefore taking the square root of all sides results in standard deviations:

$$\sqrt{1/n} \leq \frac{\sqrt{\left\langle i_b^2 \right\rangle_{\epsilon,b}}}{\sqrt{\left\langle \epsilon^2 \right\rangle_{\epsilon}}} \leq \sqrt{n}.$$

Here we also used (a bit late) the fact that the average nudge on a particular $a$ is independent of $a$ (dropping a subscript to avoid confusion that there seems to be this dependency while there is not). Note that we cannot do the same for $b$ because a particular $p(b|a)$ is given and $i_b$ is not independent of it.

One conclusion is that for growing $n$ the impact $i_b$ might become significantly larger than the nudge size, but that for small $n$ this is still manageable. I.e., for a statespace of $n=10$ we have that the impact can be maximally roughly 3 times larger than the nudge. So if the nudge size is 'small enough', also the expected impact will be small enough, as long as we stay in the 'discrete' range (not too high $n$).

# Verifying the analytical derivations of impact of nudging as function of MI

```python
epsilon = 0.001  # perturbation (nudge) size
num_fracs = 14
num_pdfs = 3
numvals_list = [2, 3, 4]
# number of nudges to be generated per pdf (makes little sense for the binary case)
# note: from 'Skewness of mixed distributions' it seems that ntrials must be at least 5k and
# preferably at least 10k as a necessary condition to measure skewness in impact and MI.
ntrials = 7000
init_cond = 'uniform'
```

```python
import time
```

```python
def denan(ls, also_inf=False):
    if hasattr(ls, '__iter__'):
        if len(ls) > 0:
            if np.ndim(ls) > 1:
                return [denan(x) for x in ls]
            elif np.ndim(ls) == 1:
                if not also_inf:
                    return [x for x in ls if not np.isnan(x)]
                else:
                    return [x for x in ls if np.isfinite(x)]
            else:
                raise ValueError()
        else:
            return ls
    else:
        return ls if not np.isnan else None
```

```python

impacts_list_nv = []  # shape (num_pdfs, num_fracs, ntrials, numvals)
nudges_list_nv = []  # shape (num_pdfs, num_fracs, ntrials, numvals)
pdfs_list_nv = []  # shape (num_pdfs, num_fracs, ntrials)
pdfs_after_list_nv = []  # shape (num_pdfs, num_fracs, ntrials)

frac_mis = np.linspace(0, 1, num_fracs)

time_init = time.time()

for numvals in numvals_list:

    # list of lists of lists of impacts (vector of probability differences)
    impacts_list = []  # shape (num_pdfs, num_fracs, ntrials, numvals)
    nudges_list = []  # shape (num_pdfs, num_fracs, ntrials, numvals)
    pdfs_list = []  # shape (num_pdfs, num_fracs, ntrials)
    pdfs_after_list = []  # shape (num_pdfs, num_fracs, ntrials)

    time_before = time.time()

    def worker(f, recurse=0):
        pdf = jp.JointProbabilityMatrix(1, numvals, init_cond)

        if f > 0:
            try:
                pdf.append_variables_with_target_mi(1, f * pdf.entropy(), 
                                                    num_repeats=int((numvals - 1) / 2) + 1)
            except UserWarning as e:
                assert 'num_repeats' in str(e)

                if recurse < 10:
                    return worker(f, recurse=recurse + 1)
                else:
                    pdf.append_independent_variables(pdf)

                    raise UserWarning(e)
        else:
            pdf.append_independent_variables(pdf)

        nudge, impact, pdf_after = pdf.susceptibility_local_single(0, 1, epsilon, 
                                                                   impact_measure='prob', 
                                                                   ntrials=ntrials,
                                                                   also_return_pdf_after=True)

        return nudge, impact, pdf, pdf_after


    for pix in xrange(num_pdfs):
        impacts = []  # shape: (len(frac_mis), ntrials, |X2|)
        nudges = []
        pdfs = []
        pdfs_after = []

        # TODO: generate multiple pdf's per fraction?

    #     for f in frac_mis:
    #         pdf = jp.JointProbabilityMatrix(1, numvals, 'uniform')
    #         if f > 0:
    #             pdf.append_variables_with_target_mi(1, f * pdf.entropy(), num_repeats=int((numvals - 1) / 2) + 1)
    #         else:
    #             pdf.append_independent_variables(pdf)

    #         nudge, impact = pdf.susceptibility_local_single(0, 1, epsilon, impact_measure='prob', ntrials=ntrials)

    #         nudges.append(nudge)
    #         impacts.append(impact)
    #         pdfs.append(pdf.copy())
        pool = mp.Pool(mp.cpu_count() - 1)
        rets = pool.map(worker, frac_mis)
        # note: the transpose somehow screws up things like np.shape(nudges_list), even
        # though the structure itself appears correct, so I replaced by the maps below...
    #     nudges, impacts, pdfs, pdfs_after = np.transpose(rets)
        nudges = map(lambda x: x[0], rets)
        impacts = map(lambda x: x[1], rets)
        pdfs = map(lambda x: x[2], rets)
        pdfs_after = map(lambda x: x[3], rets)

        impacts_list.append(impacts)
        nudges_list.append(nudges)
        pdfs_list.append(pdfs)
        pdfs_after_list.append(pdfs_after)

        if pix == 0:
            print 'note: first iteration (out of %s x %s) took %s seconds' % (num_pdfs, len(numvals_list), time.time() - time_before)

    # store all results for this numvals value
    impacts_list_nv.append(impacts_list)
    nudges_list_nv.append(nudges_list)
    pdfs_list_nv.append(pdfs_list)
    pdfs_after_list_nv.append(pdfs_after_list)
    
    print 'note: finished numvals=%s after %s seconds.' % (numvals, time.time() - time_init)
    
print 'note: entire calculation took %s seconds.' % (time.time() - time_init)
```

```python
# returns the list of specific surprises \sigma(a) for a given p(A,B) 
def specific_surprises(pdf, num_cond_vars=1):
    pdf_a = pdf.marginalize_distribution(range(num_cond_vars))
    pdf_b = pdf.marginalize_distribution(range(num_cond_vars, len(pdf)))
    return [np.sum([pdf.conditional_probability_distribution([0], a)(b) * np.log2(pdf.conditional_probability_distribution([0], a)(b) / pdf_b(b)) 
             if pdf.conditional_probability_distribution([0], a)(b) > 0 else 0.0
             for b in pdf_b.statespace()])
            for a in pdf_a.statespace()]

assert len(pdfs_list[0][0]) == 2, 'assumption implicit in num_cond_vars=1 in the specific_surprises function is that only the first variable is nudged'

# test
print specific_surprises(pdfs_list[0][5])
print np.sum(specific_surprises(pdfs_list[0][5]))
```

The above result (95th percentile of relative contribution of impact term) seems to be quite low (0.033 for numvals==4, 0.066 for numvals==3, considering all fractions), which means that the MI difference (see analytical derivation) will be only about 3% or 6% built out of the (skewed) impact term and 97% out of the (symmetric, non-skewed) correlation term. This means in turn that the skewness of the MI will practically always be very close to zero and the causal impact (presence of the impact term) cannot be distinguished in any practical dataset.

```python
rel_errors = [np.subtract(np.subtract(ct, it), md) / md 
              for ct, it, md in np.transpose([corr_terms_t, impact_terms_t, mi_diff_t], axes=[1,0,2])]
rel_errors = np.transpose(rel_errors)
```

```python
sns.plt.plot(frac_mis, map(np.mean, np.abs(rel_errors)))
sns.plt.xlabel('I(X1:X2) / H(X1)')
sns.plt.ylabel('Average absolute relative error')
sns.plt.show()

print 'note: the 95th percentile of relative error:', np.percentile(denan(map(np.mean, np.abs(rel_errors))), 95)
```

If all is well then the above plot should not have a trend (zero slope) which means that our analytical prediction is not systematically missing some MI-proportional contribution. This is important because the MI between the variables is a measure of causal strength, so if there would be trend then we would be missing something, and our method would work better either for low or for high causal relations. 

But the relative error should also not exceed 1, which means 100% error. Luckily, for numvals=4, num_pdfs=50, ntrials=50 we find a 95th percentile of absolute relative error of 0.019, which is very low. So the approximation seems successful!

```python
avg_std_rel_err = np.mean(denan(map(np.std, denan(rel_errors))))

print 'note: the average SD of the relative error is:', avg_std_rel_err, '(should be <1 and preferably <0.5, and the lower the better)'
```

Note: for epsilon=0.001 I found an average SD relative error of 0.06, for epsilon=0.01 I found an average SD relative error of 0.4, and for epsilon=0.05 I found 0.78. This supports the hypothesis that the analytical approximation gets better for smaller epsilon, which makes sense because some approximations were made in the derivations which assumed very low epsilon (ln(x) ~= x for small x).

For numvals==3 I found for epsilon=0.001 0.0505. For numvals==4 and epsilon==0.001 I get 0.031. So for increasing  numvals it also seems to become better.

```python
sns.plt.plot([0.001, 0.01, 0.05], [0.065, 0.4, 0.78])
sns.plt.xlabel('epsilon')
sns.plt.ylabel('Average SD relative error')
sns.plt.title('Hand sketch of Avg. SD of rel. errors for numvals==2, i.e.,\nhow the analytical approximation gets better for smaller nudge sizes')
# sns.plt.xscale('log')
# sns.plt.yscale('log')
sns.plt.show()
```

*Conclusion:* the analytical derivation of the impact of a nudge for causal relations based on the MI seems correct! But actually detecting causality from skewness of MI is difficult.

## So in the end, can we measure a skewness in the MI?

```python
np.shape(skew_impact_terms)
```

```python
skew_impact_terms = map(float, map(ss.skew, map(lambda a: denan(a, also_inf=True), 
                                     mi_diff_combined)))

sns.plt.plot(frac_mis, skew_impact_terms, color='0.5')
sns.plt.plot(frac_mis, [0]*len(frac_mis), '--', color='0.7')
sns.plt.xlabel('I(X1:X2) / H(X1)')
sns.plt.ylabel('Skewness of MI difference, $I\' - I$')
ylim = sns.plt.ylim()
# sns.plt.show()

skew_smooth_sigma = 5
smoothed_skews = gaussian_filter1d(map(float, skew_impact_terms), skew_smooth_sigma)
sns.plt.plot(frac_mis, smoothed_skews,
            linewidth=2)
# sns.plt.plot(frac_mis, [0]*len(frac_mis), '--k')
sns.plt.xlabel('I(X1:X2) / H(X1)')
# sns.plt.ylabel('Skewness of impact term')
sns.plt.title('Smoothing done with Gaussian kernel, $\\sigma=%s$;\nAll pdfs per fraction are combined' % skew_smooth_sigma)
sns.plt.ylim([ylim[0], min(ylim[1], np.max(smoothed_skews) * 1.2)])
sns.plt.show()

sns.distplot(skew_impact_terms[1:])
sns.plt.title('Distribution of MI difference skewness\n(mean=%s, std=%s, P(<0)=%s)' % (np.mean(skew_impact_terms[1:]), np.std(skew_impact_terms[1:]), np.less(skew_impact_terms[1:], 0).sum() / float(len(skew_impact_terms[1:]))))
sns.plt.show()
```

In the above plots we would wish to see a negative skewness (but for numvals=4, ntrials=50, num_pdfs=50 I don't see much).

### Let us look at one example result for a single pdf, which was subject to 'ntrials' nudges.

```python
# pick out a specific fraction of MI I(A:B)/H(A), should be near 1 if we are to see any skewness in impact and MI...
fix = int(num_fracs * 0.77)  

sns.plt.figure(figsize=(15, 10))
sns.plt.xlabel('Skewness of impact term')
sns.plt.ylabel('Frequency')
vals = map(float, impact_terms_list[0][fix])
bootstrap_skews = [ss.skew(np.random.choice(vals, len(vals), replace=True)) for _ in xrange(1000)]
ci_95_skew = (np.percentile(bootstrap_skews, 2.5), np.percentile(bootstrap_skews, 97.5))
sns.plt.title('Impact term distribution for frac=%s for a single pdf and ntrials=%s nudges\n(Skewness: %s, 95%% CI: %s)' % (round(frac_mis[fix], 2), ntrials, round(ss.skew(vals), 4), map(lambda x: round(x, 4), ci_95_skew)))
# sns.plt.title('Impact term distribution for frac=%s for a single pdf and ntrials=%s nudges\n(Skewness: %s)' % (round(frac_mis[fix], 2), ntrials, round(ss.skew(impact_terms_list[0][fix]), 4)))
sns.distplot(vals)
sns.plt.show()
```

Typically the above histogram *does* show a (positive) skewness, and statistically significantly so (CI does not contain zero). Hurray!

```python
# fix = int(num_fracs * 0.77)
sns.plt.figure(figsize=(15, 10))
sns.plt.xlabel('MI difference')
sns.plt.ylabel('Frequency')
vals = map(float, mi_diff_list[0][fix])
bootstrap_skews = [ss.skew(np.random.choice(vals, len(vals), replace=True)) for _ in xrange(1000)]
ci_95_skew = (np.percentile(bootstrap_skews, 2.5), np.percentile(bootstrap_skews, 97.5))
sns.plt.title('MI difference distribution for frac=%s for a single pdf and ntrials=%s nudges\n(Skewness: %s, 95%% CI: %s)' % (round(frac_mis[fix], 2), ntrials, round(ss.skew(vals), 4), map(lambda x: round(x, 4), ci_95_skew)))
sns.distplot(vals)
sns.plt.show()


# print 'note: 95 CI of skewness:', ci_95_skew
```

But the above plot typically *does not* show a negative skewness, and the CI probably contains zero. (At least for numvals=4, ntrials=50, num_pdfs=50 I got CI=[-0.26, 0.55].)

```python
# fix = int(num_fracs * 0.77)
sns.plt.figure(figsize=(15, 10))
sns.plt.xlabel('Correlation term')
sns.plt.ylabel('Frequency')
vals = map(float, np.divide(impact_terms_list[0][fix], corr_terms_list[0][fix]))
bootstrap_skews = [ss.skew(np.random.choice(vals, len(vals), replace=True)) for _ in xrange(1000)]
ci_95_skew = (np.percentile(bootstrap_skews, 2.5), np.percentile(bootstrap_skews, 97.5))
sns.plt.title('Correlation term distribution for frac=%s for a single pdf and ntrials=%s nudges\n(Skewness: %s, 95%% CI: %s)' % (round(frac_mis[fix], 2), ntrials, round(ss.skew(vals), 4), map(lambda x: round(x, 4), ci_95_skew)))
sns.distplot(vals)
sns.plt.show()
```

And above here we see the reason: in our formula "MI_diff == corr_term - impact_term" we see above that the impact term typically is 'dwarved' by the correlation term. For numvals=4 I see most values in range [-0.02, 0.02], meaning that the correlation term is often at least 50 times larger than the impact term. I.e., if we are to ever detect the skewness in MI due to the skewness of the impact term then we probably need a *humongous* dataset (i.e., ntrials must be very very high). 

```python
# print np.std(map(float, corr_terms_list[0][fix]))
# print np.mean(map(float, corr_terms_list[0][fix]))
# print np.std(map(float, impact_terms_list[0][fix]))
# print np.mean(map(float, impact_terms_list[0][fix]))
print np.std(map(float, impact_terms_list[0][fix])) / np.std(map(float, corr_terms_list[0][fix]))
```

The quotient SD(impact) / SD(correlation) is about 0.0020216 for numvals=4, epsilon=0.001. In 'Skewness for mixed distributions' you can see that this is way too low to detect the skewness of causal impact in the mutual information.

### Does the impact term i_b correlate with p(b) (b is output state)?

```python
print np.shape(impact_terms_list)
print np.shape(pdfs_list)
```

```python
# shape: (num_pdfs, num_fracs, ntrials, numvals)
impact_terms_list_terms = [[[[(np.power(i_b, 2), pdf.marginal_probability([1], [ix_b]))
#                                if pdf.marginal_probability([1], [ix_b]) > 0 
#                                else (0 if i_b == 0 else np.inf) 
                             for ix_b, i_b in enumerate(im)]
                           for im in impacts_list[pix][fix]]
                          for fix, pdf in enumerate(pdfs_list[pix])] 
                         for pix in xrange(num_pdfs)]
```

```python
np.shape(impact_terms_list_terms)
```

```python
pb_list = []
ib_list = []

for pix in xrange(num_pdfs):
    for fix, pdf in enumerate(pdfs_list[pix]):  # num_fracs
        for im in impact_terms_list_terms[pix][fix]:  # ntrials
            for i_b, p_b in im:
                pb_list.append(p_b)
                ib_list.append(i_b)
```

```python
sns.plt.plot(pb_list, ib_list, 'o', color='k', alpha=0.2)
sns.plt.xlabel('$p(b)$')
sns.plt.ylabel('$i_b^2$')
sns.plt.title('Is there a correlation between i_b and p(b) in the impact term?')
sns.plt.plot([1.0/numvals, 1.0/numvals], sns.plt.ylim(), '--')
sns.plt.plot([2.0/numvals, 2.0/numvals], sns.plt.ylim(), '--')
sns.plt.show()
```

```python
sns.plt.plot(pb_list, ib_list, 'o', color='k', alpha=0.2)
sns.plt.xlabel('$p(b)$')
sns.plt.ylabel('$i_b^2$')
sns.plt.title('Is there a correlation between i_b and p(b) in the impact term? (log-linear)')
sns.plt.plot([1.0/numvals, 1.0/numvals], sns.plt.ylim(), '--')
sns.plt.plot([2.0/numvals, 2.0/numvals], sns.plt.ylim(), '--')
sns.plt.yscale('log')
sns.plt.show()

sns.plt.plot(pb_list, ib_list, 'o', color='k', alpha=0.2)
sns.plt.xlabel('$p(b)$')
sns.plt.ylabel('$i_b^2$')
sns.plt.title('Is there a correlation between i_b and p(b) in the impact term? (log-log)')
sns.plt.plot([1.0/numvals, 1.0/numvals], sns.plt.ylim(), '--')
sns.plt.plot([2.0/numvals, 2.0/numvals], sns.plt.ylim(), '--')
sns.plt.yscale('log')
sns.plt.xscale('log')
sns.plt.show()

sns.plt.plot(pb_list, ib_list, 'o', color='k', alpha=0.2)
sns.plt.xlabel('$p(b)$')
sns.plt.ylabel('$i_b^2$')
sns.plt.title('Is there a correlation between i_b and p(b) in the impact term? (log-log, zoomed in)')
sns.plt.plot([1.0/numvals, 1.0/numvals], sns.plt.ylim(), '--')
sns.plt.plot([2.0/numvals, 2.0/numvals], sns.plt.ylim(), '--')
sns.plt.xlim([10**(-2), 1.])
sns.plt.yscale('log')
sns.plt.xscale('log')
sns.plt.show()
```

*Conclusion:* it seems that in the range [0, 1/numvals] the relation between |i_b| and p(b) is perfectly linear.
