# Testing energy functions

Gibbs measure has form

$$ \frac{\exp(e(x))}{\sum_x \exp(e(x)} $$

Let $S$ is a system with $s_i \in S := s_i \in \{0,1\}$. Proposal energy function:

$$e(s_i^{t+1}, s_i^t) = (1-s_i^{t})(s_i^{t+1}) (1 - \eta(x)) + (1-s_i^{t+1})(1 - s_i^t) \eta(x) + \mu s_i^t (1 - s_i^{t+1}) + s_i^{t} s_i^{t+1}(1 - \mu)$$

$$ e(s_i^t, s_i^{t+1}, x) = (s_i^t - 1)((2 s_i^{t+1} - 1)\eta(x) - s_i^{t+1}) + s_i^{t} (s_i^{t+1} + \mu -s_i^t \mu - \mu s_i^{t+1})$$

The Traditional SIS model is defined as follows

\begin{aligned}
\textrm{if } & s_i^t = 0 \rightarrow &p(s_i^{t+1} = 1) &= 1 - (1-\gamma)^{\sum_j s_j} \\
\textrm{else }&  & p(s_i^{t+1} = 0 | s_i^t = 1) &= \zeta
\end{aligned}

with $\zeta, \gamma \in [0, 1)$. 

# Testing with log

Start with $f\log(e(s_i, x)) = \log( \eta(s_i)(1-s_i)\sum_j s_j  + \mu(s_i) s_i))$  with $x = \sum_j s_j$

## Situation $s_i = 0$
\begin{aligned}
e^{\log(e(s_i = 0, x)} \propto 1 - (1 - \gamma)^x\\
\eta(s_i) x  \propto 1 - (1 - \gamma)^x\\
\eta(s_i) \propto \frac{1 - (1 - \gamma)^x}{x}
\end{aligned}

## Situation $s_i = 1$
\begin{aligned}
e^{\log(e(s_i = 1, x)} \propto \zeta\\
\mu(s_i) \propto \zeta\\
\end{aligned}


But this is of course a little ridiculous..

$$s_i = {0, 1} = {\textrm(sus}, \textrm{infect}}$$

\begin{equation}
\frac{\exp(e(x))}{\sum_x \exp(e(x))}

\end{equation}

\end{equation}


or in other words we obtain

\begin{aligned}
(\frac{e(0)}{e(1)})^{-1} = \frac{\zeta}{1 - \zeta} =  (\frac{e(1)}{e(0)}) = \frac{1}{1-\beta^x} - 1\\
\frac{\zeta}{1 - \zeta} + \frac{1-\zeta}{1-\zeta} = \frac{1}{1-\beta^x} 
\end{aligned}

This implies that $$\zeta = \beta^x $$

# Attempt 3

$$ e(s_i^t, s_i^{t+1}, x) = (s_i^t - 1)((2 s_i^{t+1} - 1)\eta(x) - s_i^{t+1}) + s_i^{t} (-2 s_i^{t+1} \mu + s_i^{t+1} + \mu)$$

## Case $s_i^t = 0$

If $s_i^{t+1} = 0$

$$ e(0, 0, x) = \eta(x) $$

If $s_i^{t+1} = 1$

$$e(0, 1, x) = 1 - \eta(x)$$

## Case $s_i^t = 1$

If $s_i^{t+1} = 0$
$$e(1, 0,x) = \mu$$

If $s_i^{t+1} = 1$
$$e(1, 1, x) = 1 - \mu$$

```python
eta = .2
mu  = .2

a = np.linspace(0, 1, 5)
b = a
m = Pottsis(graph = g, eta = 0,  mu = 0, \
            updateType = 'async')
results = np.zeros((a.size, b.size))
for idx, ai in enumerate(a):
    m.eta = ai
    for jdx, bi in enumerate(b):
        m.mu = bi
        m.reset()
        results[idx, jdx] = m.simulate(100).mean()
```
