```python
import sys; sys.path.insert(0, '../')
import networkx as nx, matplotlib.pyplot as plt, numpy as np
N = 100
p = .2
g = nx.duplication_divergence_graph(N, p)

fig, ax = plt.subplots(figsize = (30, 10))
colors = plt.cm.Spectral(np.linspace(0, 1, N))
pos = nx.kamada_kawai_layout(g)
nx.draw_networkx_nodes(g, pos = pos, node_color = colors)
nx.draw_networkx_edges(g, pos = pos, arrowstyle = 'fancy')
ax.axis('image')
fig.savefig('banner_example.jpg', transparent = True, bbox_inches = 'tight',\
           pad_inches = 0, dpi = 300)
fig.show()
```
