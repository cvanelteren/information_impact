```python
from Toolbox import infcy

m.r = 3.7
m.states[0] = .5
T = int(1e5)
half = int(T *.5)
res = m.simulate(T)[-half:].flatten()

ent = lambda p : -np.nansum(p * np.log(p), axis = -1 )
def binState(
             steps, \
             bins,\
             T,\
             steps):
    
    mapper     = {i : idx for idx, i in enumerate(bins)}
    rmapper    = {idx : i  for idx, i in enumerate(bins)}
    res        = m.simulate(T)
    res_binned = bins[np.digitize(res, bins)]   
    
    N = len(buffer)
    conditional = {}
    snapshots   = {}
    shape = (steps, len(mapper))
    for idx in range(steps, N):
        tmp = buffer[idx]
        
        snapshots[tmp] = snapshots.get(tmp, 0) + 1
        buff_ = conditional.get(tmp, np.zeros(shape))
        for step in range(steps):
            jdx = mapper[buffer[idx - step]]
            buff_[step, jdx] += 1
        conditional[tmp] = conditional.get(tmp, buff_)
    bRes   = binState(res_binned, steps, mapper)
    bSnap, bCon = bRes.values()
    z = sum(bSnap.values())

    zSnap = {i : j / z for i, j in bSnap.items()}
    zCon  = {i : j / bSnap.get(i)  for i, j in bCon.items()}

    for idx, (k, v) in enumerate(zCon.items()):
    #     print(v[0], zSnap.get(k))
        if idx == 0:
            Hc = np.zeros((steps))
            H  = np.zeros(len(mapper))
        Hc += ent(v) * zSnap.get(k)
        H  += v.mean(0)* zSnap.get(k)
    total = (ent(H) - Hc)/(np.log(len(mapper)))
    return dict(snapshots = snapshots, conditional = conditional, \
               mi = total)

# setup
bins       = np.linspace(0, 1, 50)

output = binState()


xi = np.arange(len(mi))
yi = total
fig = go.Figure()
bar = go.Scatter(x = xi, y  = yi)

y = np.ones(xi.size) * np.log(len(mapper))
# vlin = go.Scatter(x = np.ones(xi.size), \
#                y = y,\
#                   mode = 'lines')
fig.add_trace(bar)
# fig.add_trace(vlin)
fig.update_layout(\
                 xaxis_title = '$x^t$', \
                 yaxis_title = '$I(x^0; x^t) / I_{max}$',\
                 title = f'r = {m.r}',\
                 width = 150)
fig.show()
```
