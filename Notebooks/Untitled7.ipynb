```python
import networkx as nx, numpy as np, matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../')

from Utils import IO
df = IO.readCSV('../Graphs/Graph_min1_1.csv', header = 0, index_col = 0)
g = nx.from_pandas_adjacency(df)
from Utils.plotting import addGraphPretty
```

```python
%matplotlib notebook
fig, ax = plt.subplots()
addGraphPretty(g, ax = ax)
ax.axis('off')
fig.show()


deg = np.array([i for i in dict(g.degree(weight = 'weight')).values()])
print(deg.mean(), deg.max(), deg.min(), np.median(deg))
```
