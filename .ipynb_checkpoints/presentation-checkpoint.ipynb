```python
from matplotlib.pyplot import *
from numpy import *
n = 50
x = linspace(0, 10, n)
halfx =  linspace(0, n//2 - 1, n//2)

beta = 1
y = zeros(n)


y[:n//2]    = (1/(1 + exp(-x[:n//2] + 4)))
tmp = exp(- beta * halfx)

print(y[n//2])
y[n//2:]  = y[n//2-1] * tmp 


fig, ax = subplots()
ax.plot(x, y)
ax.axvline(x[n//2], 0, 1, color = 'red', linestyle = 'dashed')
ax.set_xlabel('Time[step]', fontsize = 15)
ax.set_ylabel('$\delta_i$', fontsize = 15)
fig.savefig('.../thesis/presentation/figures/nudge.eps')
```
