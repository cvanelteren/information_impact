```python
import networkx as nx
from matplotlib.pyplot import *
from numpy import *
graph = nx.krackhardt_kite_graph()

tmp = []
for f in [nx.degree, nx.closeness_centrality, nx.betweenness_centrality]:
    tmp.append(dict(f(graph)))
    
x = linspace(0, 1, graph.number_of_nodes())
colors = cm.viridis(x)
# print(colors)
pos = nx.spring_layout(graph)
for i in tmp:
    fig, ax = subplots()
    x = array(list(i.values()))
    idx = argsort(x)
#     print(x[idx])
    c = colors[idx, :]
    print(x, idx)
#     nx.draw_networkx_nodes(graph, pos, ax = ax, node_color = c, with_label = True)
    nx.draw(graph, pos, ax = ax, node_color = c, with_labels = True)
```
