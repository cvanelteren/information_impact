```python
from scipy import optimize
from numpy import *
from matplotlib.pyplot import *

func = lambda x, a, b, c: a +  b * exp(-c * x)
x = linspace(-1, 5, 100)
coeffs_real = (0, .01, .3)
y = func(x, *coeffs_real) + random.randn(len(x)) * 1e-6
coeffs, _ = optimize.curve_fit(func, x, y)

theta   = 1e-3
l       = theta
l       = coeffs[0] + theta * (coeffs[0] + coeffs[1])
root    = -1/coeffs_real[2] * (log((l - coeffs_real[0])) - log(coeffs_real[1])) # tmp

fig, ax = subplots()
ax.scatter(x, y)
ax.plot(x, func(x, *coeffs), 'r')
ax.scatter(root, func(root, *coeffs), color ='g', s = 1000)
# ax.set_xlim(x.min(), x.max())
print(theta, root, func(root, *coeffs), l)
show()
```
